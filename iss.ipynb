```python
''' Data Collection from the ISS '''

import requests
import json
import time

api = 'https://api.wheretheiss.at/v1/satellites/25544'
desiredPoints = 500

ISS_DataXY = []
ISS_DataTime = []

for i in range(desiredPoints):
    request = requests.get(api).json()
    longitude = request['longitude']
    timestamp = request['timestamp']
    latitude = request['latitude']
      
    ISS_DataTime.append([timestamp])
    ISS_DataXY.append([longitude, latitude])

    if (i+1)%(0.25*desiredPoints) == 0 and (i+1) != desiredPoints:
        progress = int(((i+1)/desiredPoints)*100)
        print (f"{progress}% of points retrieved")
        time.sleep(10)
    else:
        time.sleep(10)

print("100% of Desired Points Retrieved")
print("ISS Data Gathering Complete.")
```

```python
''' Visual Results of the Program '''

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Add the background
background = mpimg.imread("lat-lon.jpg")
fig, ax = plt.subplots()
ax.imshow(background, extent=[-180,180,-90,90])

# Plot the Points
x = [i[0] for i in ISS_DataXY]
y = [i[1] for i in ISS_DataXY]

plt.plot(x[0], y[0], "o", ms = 7, mfc = "#00ffee", label = "START of Recording")
plt.plot(x[0:desiredPoints-2], y[0:desiredPoints-2], ".r", ms = 1, label = "Path")
plt.plot(x[desiredPoints-1], y[desiredPoints-1], "o", ms = 7.5, mfc = "#f3ff00", label = "END of Recording")

# Details + Legend
plt.title("ISS Path Over 500 Points")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(bbox_to_anchor = (1.05, 1), loc = "upper left", title = "LEGEND\n")
plt.show()
```

```python
''' Display Coordinates '''

print("Below are the coordinates used to plot the path of the ISS Above:\n")
for i in range (desiredPoints):
    print (f"{i+1}. {ISS_DataTime[i]} -- Coordinates: ({ISS_DataXY[i][0]}, {ISS_DataXY[i][1]})")
```
